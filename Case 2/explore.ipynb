```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# import data 
data = pd.read_csv('HR_data.csv', index_col=0)
```

```python
# split data in emotions and measures
emotions = ['Frustrated',
       'upset', 'hostile', 'alert', 'ashamed', 'inspired', 'nervous',
       'attentive', 'afraid', 'active', 'determined']
categories = ['Round','Phase','Individual', 'Puzzler','Cohort']



# split data in emotions and measures
categoriesemotions = categories + emotions
data_emotions = data[categoriesemotions]
data_measures = data.drop(columns=emotions)

```

### Correlation matrix

```python
# Correlation matrix
corr_emotions = data_emotions[emotions].corr()
corr_measures = data_measures.drop(columns=categories).corr()
```

```python
plt.matshow(corr_emotions)
plt.title('Correlation matrix of emotions')
plt.xticks(range(len(emotions)), emotions, rotation=90)
plt.yticks(range(len(emotions)), emotions)
plt.show()
```

```python
plt.matshow(corr_measures)
labels = data_measures.drop(columns=categories).columns
plt.title('Correlation matrix of measures')
plt.xticks(range(len(labels)), labels, rotation=90)
plt.yticks(range(len(labels)), labels)
plt.show()
```

### Measures


```python
mean_measures = ['HR_TD_Mean','TEMP_TD_Mean','EDA_TD_P_Mean','EDA_TD_T_Mean']
categories_mean_measures = categories + mean_measures
data_mean_measures = data[categories_mean_measures]
```

```python
# plot average measures grouped by round
data_mean_measures.drop(columns=['Individual','Puzzler']).groupby('Round').mean(numeric_only=True).plot(kind='bar')
plt.title('Average measures by round')
plt.ylabel('Average measure')
# place legend box outside of plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# plot average measures grouped by phase
data_mean_measures.drop(columns=['Individual','Puzzler']).groupby('Phase').mean(numeric_only=True).plot(kind='bar')
plt.title('Average measures by phase')
plt.ylabel('Average measure')
# place legend box outside of plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# plot average measures grouped by cohort
data_mean_measures.drop(columns=['Individual','Puzzler']).groupby('Cohort').mean(numeric_only=True).plot(kind='bar')
plt.title('Average measures by cohort')
plt.ylabel('Average measure')
# place legend box outside of plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
# plot average measures grouped by puzzler or not
data_mean_measures.drop(columns=['Individual']).groupby('Puzzler').mean(numeric_only=True).plot(kind='bar')
plt.title('Average measures by puzzler or not')
plt.ylabel('Average measure')
# place legend box outside of plot
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
